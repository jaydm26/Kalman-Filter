# Extended Kalman Filter

The Extended Kalman Filter (EKF) is an extension to the Kalman Filter in that it allows for $A$ to be non-linear in the state-space model.

Let us look at a case where A is non-linear and see how a standard Kalman Filter performs.

Let $A$ depend on $x$ such that

$\dot{x_t} = f(x_t, u_t) + \epsilon_x$

Let $f(x, u) = x^2$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def feed_forward(t, x):
    return -x * x
```

```python
sol1 = solve_ivp(feed_forward, [0, 1], [2], method = 'BDF')
```

```python
plt.plot(sol1.t, sol1.y[0])
```

```python
np.random.seed(1234)
t1 = np.linspace(0, 1, 51)
dt = t1[1] - t1[0]
x1 = np.zeros(t1.size)
x1[0] = 2
for i in range(1,t1.size):
    x1[i] = -x1[i-1] * x1[i-1] * dt + x1[i-1] + np.random.randn() * 0.1 + np.random.randn() * 0.1
```

```python
plt.plot(sol1.t, sol1.y[0], 'b', label = 'Noise-free Solution')
plt.plot(t1, x1, 'r.', label = 'Noisy Solution')
plt.plot(t1, 1/(0.5+t1), 'k', label = 'Real Solution')
plt.legend()
plt.show()
```

This just shows that a simple Euler (BDF1) is not suitable to solve this problem. Of course, a higher order BDF (or RK45) will be able to solve this problem.

The problem, although trivial in nature, is non-linear as order of the polynomial equation on the right hand side is not 1.

Should we attempt to use a Kalman Filter on this setup, we will run into a problem when we have to obtain the derivative of $f(x)$. While in this trivial case, the derivative is easy to calculate, the problem serves as an illustration to the technique of implementing Extended Kalman Filters.

```python
t = np.linspace(0, 1, 51)
dt = t[1] - t[0]
x = np.zeros(t.size) # observed
x_hat_prime = np.zeros(t.size) # predicted state (prior estimate)
x_hat = np.zeros(t.size) # posterior estimate of theta_hat_prime
x_real = np.zeros(t.size) # real theta including noise
P = np.zeros(t.size) # Posterior estimate of error covariance
P_prime = np.zeros(t.size) # Prior estimate of error covariance
K = np.zeros(t.size) # Kalman Gain
P_prime[0] = 2
x[0] = 2
x_hat_prime[0] = 2
x_hat[0] = 2
x_real[0] = 2
```

Steps within a Extended Kalman Filter Algorithm are as follows:

1. Calculate Kalman Gain $K_k = P_k'C^T(CP_k'C^T + R)^{-1}$ where $P_k'$ is the prior estimate of the error, $R$ is the covariance matrix of the observation noise.
2. Update the prediction using $\hat{x} = \hat{x}' + K_k(z_k - C\hat{x}')$, where $z_k$ is the observed measurement.
3. Update the estimate of the error $P_k = (I - K_k C) P_k'$
4. Move to the next time step through a projection step:
$\hat{x}_{k+1}' = f(x_k) $,  
$P_{k+1}' = F P_k F^T + Q$,
where $Q$ is the covariance matrix of the model noise, and $F$ is the Linearised Taylor Series Expansion for the first deriviate of $f(x)$ about $x$.

```python
C = 1
f = lambda x: -x * x # This could be a highly complex motion dynamics equation.
f_prime = lambda x : (f(x+0.01) - f(x)) * 100 # Taylor Series Expansion about x
# Pro-tip: Use an automatic differentiation package (Tensorflow, PyTorch, etc.) for
# immediate gradient calculations about a point.
```

```python
np.random.seed(1234)
for i in range(t.size - 1):
    K[i] = P_prime[i] * C / (C * P_prime[i] * C) + 0.01 # No noise is present in the current model.
    x_hat[i] = x_hat_prime[i] + K[i] * (x[i] - C * x_hat_prime[i])
    P[i] = (1 - K[i] * C) * P_prime[i]
    x_hat_prime[i+1] = (f(x_hat[i]) * dt) + x_hat[i]
    x_real[i+1] = x_hat_prime[i+1] + np.random.randn() * 0.1
    x[i+1] = C * x_real[i+1] + np.random.randn() * 0.1
    P_prime[i+1] = f_prime(x_real[i]) * P[i] * f_prime(x_real[i]) + 0.01
    
i = t.size-1
K[i] = P_prime[i] * C / (C * P_prime[i] * C) + 0.01
x_hat[i] = x_hat_prime[i] + K[i] * (x[i] - C * x_hat_prime[i])
P[i] = (1 - K[i] * C) * P_prime[i]
```

```python
plt.plot(t, x, 'rx', label = 'Observed')
plt.plot(t, x_hat, 'g', label = 'Posterior Noise-free State')
# plt.plot(t, x_hat_prime, 'm', label = 'Prior Noise-free State/Model Prediction' )
plt.plot(t, 1/(0.5+t), 'b', label = 'Real Solution')
# plt.plot(t1, x1, 'k+', label = 'From Previous')
plt.legend()
plt.show()
```
